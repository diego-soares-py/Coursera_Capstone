```python
#!pip install geocoder

import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
#import geocoder # import geocoder

url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = urllib.request.urlopen(url)

soup = BeautifulSoup(page, "lxml")
all_tables=soup.find_all("table")

!wget -O Geospatial_Coordinates.csv https://ibm.box.com/shared/static/9afzr83pps4pwf2smjjcf1y5mvgb18rr.csv

df_c = pd.read_csv('Geospatial_Coordinates.csv' )
```

```python
right_table=soup.find('table', class_='wikitable sortable')
df_c = df_c.rename(columns={'Postal Code':'Postal_Code'})
df_c = df_c.set_index(['Postal_Code'])

    
```

```python
def get_latitude(postal_code):
    latitude = '0'
    #latitude = str(df_c.at[df_c['Postal Code'].eq(postal_code), 'Latitude'])
    latitude = df_c.loc[postal_code]['Latitude']
    print(latitude)
    return latitude

def get_longitude(postal_code):
    longitude = '0'
    #longitude = df_c.at[df_c['Postal Code'].eq(postal_code), 'Longitude']
    #longitude = df_c.loc[df_c['Postal Code'] == postal_code]
    #df_c.loc[df_c['Postal Code'] == 30,'Height']
    longitude = df_c.loc[postal_code]['Longitude']
    print(longitude)
    return longitude

```

Clear and transform data

```python
A=[]
B=[]
C=[]
D=[]
E=[]


for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        zip_code= cells[0].find(text=True)
        aux = cells[1].find(text=True)
        aux2 = cells[2].find(text=True)
        zip_code = zip_code.replace('\n' , '')
        #print(aux)
        if "assigned" not in aux: 
            A.append(zip_code)
            B.append(cells[1].find(text=True))
            if "assigned" not in aux2:
                C.append(cells[2].find(text=True))
            else:
                C.append(cells[1].find(text=True))
            print(zip_code)
            D.append(get_latitude(zip_code))
            E.append(get_longitude(zip_code))
        

df=pd.DataFrame(A,columns=['Postal Code'])
df['Borough']=B
df['Neighborhood']=C
df['Latitude']=D
df['Longitude']=E

```
